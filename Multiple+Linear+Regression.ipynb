```python
import pandas as pd
import numpy as np

nasdaqFile = 'C:/Users/efran/Documents/Data/IXIC.csv'
spFile = 'C:/Users/efran/Documents/Data/GSPC.csv'
googFile = 'C:/Users/efran/Documents/Data/GOOG.csv'
oilFile = 'C:/Users/efran/Documents/Data/USO.csv'
xomFile = 'C:/Users/efran/Documents/Data/XOM.csv'
```

```python
def readFile(filename):
    data=pd.read_csv(filename,sep=",",usecols=[0,5],names=['Date', 'Price'],header=0)
    returns=np.array(data["Price"][:-1],np.float)/np.array(data["Price"][1:],np.float)-1
    data["Returns"]=np.append(returns,np.nan)
    data.index=data["Date"]
    return data
```

```python
from sklearn.linear_model import SGDRegressor,LinearRegression
```

```python
oilData = readFile(oilFile)
nasdaqData = readFile(nasdaqFile)
xomData = readFile(xomFile)
```
